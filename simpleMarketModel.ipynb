```python
from agent import Agent
from market import Market
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(rc={'figure.figsize':(16, 5)})
```

```python
m=Market(500, 0.5)

WTP_ = [a.w2p for a in m.agents_ if a.role == 2]
WTA_ = [a.w2a for a in m.agents_ if a.role == 1]

WTP = np.array(sorted(WTP_, reverse=True))
WTA = np.array(sorted(WTA_))
```

```python
max_index = min(len(WTP), len(WTA))
wtp=WTP[:max_index]
wta=WTA[:max_index]
diffs=abs(wtp-wta)
min_diff=min(diffs)

Q=np.where(diffs==min_diff)
P=wta[Q]

_ = plt.plot(np.arange(0, len(WTP)), WTP, color='blue')
_ = plt.plot(np.arange(0, len(WTA)), WTA, color='orange')
_ = plt.title("Quantity sold: {q} at equilibrium price {p}".format(q=Q[0][0], p=np.trunc(P[0]*100)/100))
_ = plt.margins(x=0.01, y=0.05)
```

```python
aVer = []
for i in range(1000):
    x=np.mean(m.random_q())
    aVer.append(x)
    
    print('running at:', i)
```

```python
_ = plt.plot(np.arange(500, len(aVer)), aVer[500:])
```

```python
m.plot_pricesCDF(aVer)
```
